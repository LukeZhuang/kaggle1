```python
import numpy as np
import pymysql
import time

num_train=260864
num_test=261477
```

```python
conn = pymysql.connect(host='localhost', port=3306,user='root',passwd='root',db='cse258',charset='UTF8')
cur = conn.cursor()
```

```python
def calc_feature(user):
    cmd='SELECT COUNT(*)FROM user_log WHERE user_id='+user+' GROUP BY SUBSTR(time_stamp,1,LENGTH(time_stamp)-2)'
    cur.execute(cmd)
    result=cur.fetchall()
    result=np.array(result)
    return [result.shape[0],np.sum(result),np.mean(result),np.std(result)]
```

```python
user_info={}

X=[]
f_train=open('data_format1/train_format1.csv')
cnt=0
print_every=5000
for line in f_train:
    l=line.strip()
    if l=='user_id,merchant_id,label':
        continue
    [user,seller,this_y]=l.split(',')
    
    if user not in user_info:
        res=calc_feature(user)
        user_info[user]=res
    else:
        res=user_info[user]
    X.append([res[0],res[1],res[2],res[3]])
    
    if cnt==0 or (cnt+1)%print_every==0 or (cnt+1)==num_train:
        print(time.strftime('%Y-%m-%d %H:%M:%S',time.localtime(time.time())),
              'finished '+str(cnt+1)+' pairs')
    cnt+=1

X=np.array(X)
print(X.shape)
print(X)
np.save('data/x_monthly.npy', X)
```

```python
X_test=[]
f_test=open('data_format1/test_format1.csv')
cnt=0
print_every=5000
for line in f_test:
    l=line.strip()
    if l=='user_id,merchant_id,prob':
        continue
    [user,seller,_]=l.split(',')
    
    if user not in user_info:
        res=calc_feature(user)
        user_info[user]=res
    else:
        res=user_info[user]
        
    X_test.append([res[0],res[1],res[2],res[3]])
    
    if cnt==0 or (cnt+1)%print_every==0 or (cnt+1)==num_test:
        print(time.strftime('%Y-%m-%d %H:%M:%S',time.localtime(time.time())),
              'finished '+str(cnt+1)+' pairs')
    cnt+=1

X_test=np.array(X_test)
print(X_test.shape)
print(X_test)
np.save('data/x_monthly_test.npy', X_test)
```
